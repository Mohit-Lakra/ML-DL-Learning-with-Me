```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine curve')
plt.plot()
```

```python
# parabola
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'bx')
plt.show()
```

```python
x = np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'g.')
plt.plot(x,np.cos(x),'r')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['English','Hindi','Tamil','Punjabi']
people = [100,500,200,200]
ax.bar(lang,people)
plt.xlabel('languages')
plt.ylabel('number of people')
plt.title('languages spoken')
plt.show()
```

```python
fig1 = plt.figure();
ax = fig1.add_axes([0,0,1,1])
lang = ['English','Hindi','Tamil','Punjabi']
people = [100,500,200,200]
ax.pie(people,labels=lang,autopct='%1.1f%%')
plt.xlabel('languages')
plt.ylabel('number of people')
plt.title('languages spoken')
plt.show()
```

```python
x = np.linspace(-5,5,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

```python
#3D scatter plot
fig3 = plt.figure()
z = 20 * np.random.random(1000)
x = np.random.randn(1000)
y = np.random.randn(1000)
ax = fig3.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c='r',marker='o')
plt.show()
```
